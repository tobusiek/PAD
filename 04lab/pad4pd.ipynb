Zad 4

```python
zad4 = pd.read_csv('Zadanie_4.csv', sep=';')
zad4
```

```python
zad4 = zad4.loc[:, ~zad4.columns.str.contains('^Unnamed')]
zad4
```

```python
zad4['Date'] = pd.to_datetime(zad4['DateTime']).dt.date
zad4['Time'] = pd.to_datetime(zad4['DateTime']).dt.time
zad4
```

```python
zad4.drop(labels='DateTime', inplace=True, axis=1)
zad4
```

```python
df1 = zad4.groupby(['Date', 'City', 'DoctorID', 'Type']).min()
df1.reset_index()
df1
```

```python
df2 = zad4.groupby(['Date', 'City', 'DoctorID', 'Type']).max()
df2.reset_index()
df2
```

```python
df3 = pd.merge(df1, df2, how='inner', on=['Date', 'City', 'DoctorID', 'Type'])
df3
```

```python
df3.rename(columns={'Time_x':'From', 'Time_y':'To'}, inplace=True)
df3.reset_index()
df3
```

```python
df3.to_csv('doc_availability.csv', sep=';')
```

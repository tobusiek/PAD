```python
import pandas as pd

# zad1
df = pd.read_csv('PAD_03_PD.csv', sep=';', index_col='ID')
df
```

```python
# np.where(pd.isnull(df))

for attr in df.columns:
    empty_records = df[pd.isna(df[attr]) == True]
    empty_records_idxs = empty_records.index.tolist()
    df.drop(empty_records_idxs, inplace=True)
        # df.drop(axis=0, index=record.index, inplace=True)
```

```python
df['Country'].value_counts()
```

```python
# zad2
df['owned_goods'] = df['owns_car'] + df['owns_TV'] + df['owns_house'] + df['owns_Phone']
df
```

```python
# zad3
import numpy as np

goods_avg_balance = np.mean(df['gender'] == 'K') / np.mean(df['gender'] == 'M')
goods_avg_balance.round(2)
```

```python
# zad4
countries = pd.unique(df['Country'])
countries
```

```python
owned_goods_avg_in_country = df.groupby('Country')['owned_goods'].mean().round(2).values
owned_goods_avg_in_country
```

```python
min_age_in_countries = df.groupby('Country')['Age'].min().values
min_age_in_countries
```

```python
new_df = pd.DataFrame({'Country': countries, 'Avg_goods_owned': owned_goods_avg_in_country,
                       'Min_age': min_age_in_countries})
new_df
```

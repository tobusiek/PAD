# 1. Wprowadzenie do Pandas

## Pandas

- Pandas, wraz z Numpy, są prawdopodobnie najważniejszą biblioteką Python do Data Science.
- Podstawowym obiektem Pandas jest DataFrame, są jeszcze Series ale nie będziemy ich omawiać na tym przedmiocie.
- Najlepszą metaforą dla Pandas, jest wyobrażenie sobie przeogromnej i potężnej wersji Excela, dla którego DataFrame jest akruszem.
- DataFrames ma wiersze i kolumny.
- Każda kolumna DataFrame jest obiektem Pandas Series.
- Obiekt Pandas series jest zbudowany z listy (array) Numpy (Series jest jak array lista ale z otagowanymi indeksami).
- Zgodnie z konwencją importujemy pandas z aliasem:

```python
import pandas as pd
```

```python
# na razie nie przejmuj się danymi, wczytujemy plik csv do pandas w celu demonstracji operacji
sf_sal = pd.read_csv('Salaries.csv')
sf_sal
```

```python
# tworzymy sobie listę indeksów składających się z dwóch liter. Jest to tylko POMOCNICZE

import string
alphabet = string.ascii_uppercase
alphabet

index_list = []

for first in alphabet:
    for second in alphabet:
            index_list.append(first + second)
```

```python
print(index_list)
```

```python
sf_sal["Id"] = index_list
sf_sal.set_index("Id", inplace=True)
sf_sal
```

## Selection i Indexing

- Aby indeksować dane używamy nawiasów kadratowych, głównie dla kolumn.
- Możemy również przekazać listę kolumn do wyboru.
- Pojedyncze nawiasy kwadratowe zwrócą obiekt Series, a podwójne DataFrame.
- Sugeruję używać podwójne nawiasy, o ile nie potrzebujemy obiektu Series.

```python
# pojedyncze nawiazy zwracają Series, wraz z nazwą/długością/informacjami o typie danych
# pamiętaj, że można je uzyskać bezpośrednio za pomocą metod

sf_sal["BasePay"]
```

```python
# podwójne nawiasy zwracają DataFrame

sf_sal[["BasePay"]]
```

```python
# aby wybrać więcej niż jedną kolumnę potrzeba podwójnych nawiasów

sf_sal[["BasePay","TotalPay"]]
```

## NoneType
- Musimy tutaj nauczyć się rozróżniać None, NaN i 0.
- None jest typem 'NoneType' i dlatego jest wartością, można go wykorzystywać jako symbol zastępczy przed dodaniem wartości.
- 0 jest integerem, a zatem wartością. Oznacza to, że komórka w tabeli nie jest pusta, ale ma wartość 0.
- NaN oznacza 'Not a Number' i wskazuje na BRAKUJĄCĄ wartość, pomimo, że jest typu float.
- Możemy to lepiej dostrzec sprawdzając każdy z typów:

```python
type(None)
```

```python
type(0)
```

```python
import numpy as np

type(np.nan)
np.nan + float("inf")
```

## Dodawanie i usuwanie kolumn

```python
# nową kolumnę można utworzyć pdając jedną wartość, włączając integer/string/NoneType

sf_sal["new col"] = None

sf_sal.head()
```

```python
# można usuwać wiersze/kolumny metodą .drop(), axis=0 dla wierszy, axis=1 dla kolumn
# inplace zapisujące wynik metody z defaultu jest na False, dlatego wynik wyświetli DataFrame bez usuniętych wierszy/kolumn, ale nie usunie ich z obiektu

sf_sal.drop(["AE", "AB"]) # axis=0, inplace=False by default
```

```python
sf_sal
```

```python
sf_sal.drop("new col", axis=1, inplace=True)

sf_sal.head()
```

## Loc and Iloc

```python
# do zwracania wierszy po nazwie wiersza i kolumn po nazwie kolumny korzysta się z .loc[]

sf_sal.loc["AA"]
```

```python
# do zwracania wierszy po indeksie używa się .iloc[rows,columns] (jak zawsze zaczyna się od 0)

sf_sal.iloc[0]
```

```python
# aby zwrócić pojedynczą komórkę można podać .loc[r,c] z wierszem i kolumną

sf_sal.loc["AL", "TotalPay"]
```

```python
# do zwrócenia podzbioru wierszy i kolumn można wykorzystać lity w .loc[]
# może indeksować nie w kolejności

sf_sal.loc[["AC", "CV", "BK"],["BasePay", "Surname", "Year"]]
```

## Wybór warunkowy

```python
# warunek logiczny taki jak ten zwraca boolean zastosowany do każdej wartości w kolumnie (np. NumPy)

sf_sal["TotalPay"] > 300000
```

```python
# może indeksować DataFrame za pomocą tej wartości logicznej, aby zwrócić tylko wiersze, w których jest to prawda (tzw. maskowanie wartości logicznej)

mask = sf_sal["TotalPay"] > 300000

sf_sal_mask = sf_sal[mask]
sf_sal_mask
```

```python
# z dużą ramką DataFrame, znacznie bardziej wydajnym pamięciowo sposobem indeksowanie jest przykład poniżej
# aby uniknąć zapisywania ogromnej serii logicznej do zmiennej bez powodu

sf_sal[sf_sal["TotalPay"] > 300000]
```

```python
# podzbiór można dalej indeksować

sf_sal[sf_sal["TotalPay"] > 300000][["JobTitle", "Year", "TotalPay"]]
```
